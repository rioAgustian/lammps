```python
import numpy as np
import matplotlib.pyplot as plt

file = r"C:\Users\Rio Agustian\AppData\Local\LAMMPS 64-bit 12Jun2025\DATA\lennard-jones-fluid\number_data.txt"

# Data
data = np.loadtxt(file, skiprows=1)
steps = data[:, 0] * 0.005
temp  = data[:, 1] 
v_n1_in = data[:, 6]
v_n2_in = data[:, 7]
coor12  = data[:, 8] * 100

plt.figure(figsize=(13, 5))

plt.subplot(121)
plt.plot(steps, v_n1_in, color='orange', label='N1_in')
plt.plot(steps, v_n2_in, color='blue', label='N2_in')
plt.xlabel("t")
plt.ylabel("N_in")
plt.legend()

plt.subplot(122)
plt.plot(steps, coor12, color='cyan')
plt.xlabel("t")
plt.ylabel("C12")

plt.tight_layout()  # Adjust layout to fit suptitle
plt.savefig("my_plot.png")
plt.show()
```
